# Reflect Tables into SQLAlchemy ORM

```python
# Python SQL toolkit and Object Relational Mapper
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, func, inspect
```

```python
engine = create_engine("sqlite:///Resources/hawaii.sqlite")
```

```python
# reflect an existing database into a new model
Base = automap_base()
# reflect the tables
Base.prepare(engine, reflect=True)
```

```python
# We can view all of the classes that automap found
Base.classes.keys()
```

```python
# Save references to each table
Measurement = Base.classes.measurement
Station = Base.classes.station
```

```python
# Create our session (link) from Python to the DB
session = Session(engine)
```

```python
#Listing all of the precipitation values and date in the measurement table
session.query(Measurement.date, Measurement.prcp).distinct().all()
```

```python
# finding the latest precipitation date
latest = session.query(Measurement.date).order_by(Measurement.date.desc()).first()
latest
```

```python
#extracting string from query object
latest = list(np.ravel(latest))[0]

#converting date string to datetime object
latest = dt.datetime.strptime(latest, '%Y-%m-%d')

#extracting year, month, and day as integers
latestYear = int(dt.datetime.strftime(latest, '%Y'))
latestMonth = int(dt.datetime.strftime(latest, '%m'))
latestDay = int(dt.datetime.strftime(latest, '%d'))
```

```python
# Calculate the date 1 year ago from the last data point in the database

yearago = dt.date(latestYear, latestMonth, latestDay) - dt.timedelta(days=365)

#query for dates and precipitation for the latest year
latest_precip = (session.query(Measurement.date, Measurement.prcp)
                  .filter(Measurement.date > yearago)
                  .order_by(Measurement.date)
                  .all())
```

```python
#convert query object to data frame and set index to "date" (helps on plotting)
precip_df = pd.DataFrame(latest_precip)
```

```python
precip_df = precip_df.fillna(0)
precip_df = precip_df.set_index('date')
```

```python
#sort by date
precip_df = precip_df.sort_index(ascending=True)
precip_df = precip_df.rename(columns = {"prcp": "precipitation"}) 
```

```python
# Use Pandas Plotting with Matplotlib to plot the data
precip_df.plot(title = "Precipitation: 8/23/16 - 8/23/17",
               color='b', 
               alpha=0.5,
               rot = '45')

plt.legend(loc='upper center')
plt.savefig("Images/Precipitation01.png")
plt.show()
```

```python
# Use Pandas to calcualte the summary statistics for the precipitation data
precip_df.describe()
```

```python
# Design a query to show how many stations are available in this dataset?
count_stations = session.query(Station.id).count() 

print(f'There are {count_stations} weather stations available in this dataset.')
```

```python
# What are the most active stations? (i.e. what stations have the most rows)?
# List the stations and the counts in descending order.
# finds station activity and sorts from most active to least active.

station_activity = session.query(Measurement.station, Station.name, func.count(Measurement.tobs)).\
filter(Measurement.station == Station.station).group_by(Measurement.station).order_by(func.count(Measurement.tobs).desc()).all()
```

```python
station_activity
```

```python
print(f"The most active station is {station_activity[0][1][0:6]}.")
```

```python
# Using the d from the previous query, calculate the lowest temperature recorded, 
# highest temperature recorded, and average temperature most active station?



stationID = station_activity[0][0] #getting ID from previous query

#query for highest temperature
high_Temp = (session.query(Measurement.tobs)
                      .filter(Measurement.station == stationID)
                      .order_by(Measurement.tobs.desc())
                      .first())
print(f"The highest temperature recorded there is {high_Temp[0]}F.")
```

```python
#query for lowest temperature
low_Temp = (session.query(Measurement.tobs)
                     .filter(Measurement.station == stationID)
                     .order_by(Measurement.tobs.asc())
                     .first())

print(f"The lowest temperature recorded there is {low_Temp} F.")
```

```python
#query for average temperature
avg_Temp = (session.query(func.avg(Measurement.tobs))
                  .filter(Measurement.station == stationID))

avg_Temp = '{0:.3}'.format(avg_Temp[0][0])
print(f"The average temperature recorded there is {avg_Temp} degrees Farenheit.")
```

```python
# Choose the station with the highest number of temperature observations.
# Query the last 12 months of temperature observation data for this station and plot the results as a histogram

    
#query for the last year of temperature data
temp_Data = (session.query(Measurement.date, Measurement.tobs)
                   .filter(Measurement.date > yearago)
                   .filter(Measurement.station == stationID)
                   .order_by(Measurement.date)
                   .all())

#converting to data frame
temp_df = pd.DataFrame(temp_Data)
temp_df = temp_df.set_index('date')

#sorting by date
temp_df = temp_df.sort_index(ascending=True)
temp_df = temp_df.rename(columns = {"tobs": "temperature"}) 
```

```python
#histogram plot with pandas
temp_df.plot(kind='hist', bins=12)    
plt.title(f"Temperature Observation for {station_activity[0][1][0:6]}")

plt.legend(loc='upper left')
plt.xlabel('temperature (F)')
plt.savefig("Images/stationhistog01.png")
plt.show()
```

```python
# This function called `calc_temps` will accept start date and end date in the format '%Y-%m-%d' 
# and return the minimum, average, and maximum temperatures for that range of dates

def calc_temps(start_date, end_date):
    temp_list = []
    
    results = session.query(Measurement.tobs).filter(Measurement.date>=start_date, Measurement.date<=end_date).all()
    
    temps = [temp[0] for temp in results]
    
    temp_list.append(np.mean(temps))
    temp_list.append(min(temps))
    temp_list.append(max(temps))
    
    return temp_list
```

```python
# function usage example
print(calc_temps('2012-02-28', '2012-03-05'))
```

```python
# Use your previous function `calc_temps` to calculate the tmin, tavg, and tmax 
# for your trip using the previous year's data for those same dates.
print(calc_temps('2011-02-28', '2011-03-05'))
```

```python
#set trip dates and previous dates

trip = '2019-06-20 to \n 2019-06-27'

trip_start = '2017-05-08'
trip_end = '2017-05-25'

#calculate the tmin, tavg, and tmax 
trip_temps = calc_temps(trip_start, trip_end)
trip_temps
```

```python
# Plot the results from your previous query as a bar chart. 
# Use "Trip Avg Temp" as your Title
# Use the average temperature for the y value
# Use the peak-to-peak (tmax-tmin) value as the y error bar (yerr)

plt.figure(figsize=(3,5))

plt.bar(1, trip_temps[0], 
        yerr=(trip_temps[2] - trip_temps[1]), 
        color='salmon')

plt.ylabel("Temperature (F)")
plt.xlabel("")
plt.title(f"Avg Trip Temp")

plt.savefig("Images/temperatures.png")
plt.show()
```

```python
# Calculate the total amount of rainfall per weather station for your trip dates using the previous year's matching dates.
# Sort this in descending order by precipitation amount and list the station, name, latitude, longitude, and elevation


station_rain = session.query(Measurement.station, Station.name, Station.latitude, Station.longitude, Station.elevation, func.sum(Measurement.prcp).label('precipitation')).\
filter(Measurement.station == Station.station, Measurement.date >= trip_start, Measurement.date <= trip_end).\
group_by(Measurement.station).order_by(func.sum(Measurement.prcp).desc()).all()


print(station_rain)
```

```python
# Create a query that will calculate the daily normals 
# (i.e. the averages for tmin, tmax, and tavg for all historic data matching a specific month and day)
def daily_normals(date):
        """Daily Normals.
        """
        sel = [func.min(Measurement.tobs), func.avg(Measurement.tobs), func.max(Measurement.tobs)]
        return session.query(*sel).filter(func.strftime("%m-%d", Measurement.date) == date).all()

daily_normals("01-01")
```

```python
# I created a list of dates for trips

trip_dates = ["01-04", "01-05", "01-06", 
              "01-07", "01-08", "01-09", 
              "01-10", "01-11", "01-12", 
              "01-13", "01-14","01-15", 
              "01-16", "01-17", "01-18",
             "01-19", "01-20", "01-21",
             "01-22", "01-23", "01-24",
             "01-25", "01-26", "01-27"]

#calling daily normals and calculate normals for each date
daily_norms = [daily_normals(date) for date in trip_dates]

daily_norms
```

```python
#extract normals into a list of lists and convert to a dataframe
normals = [np.array(normal[0]) for normal in daily_norms]

normals_df = pd.DataFrame(normals)
normals_df.rename(columns={0: "Low Temp", 1: "Avg Temp", 2: "High Temp"})
```

```python
normals_df.plot(figsize=(12, 6),kind='area',sort_columns=True,use_index=True, title='Daily Normals', 
                stacked=False, fontsize=20,grid=True) 

plt.xlabel('date', fontsize=16)
plt.ylabel('Temperature (F)', fontsize = 16)

plt.legend(['low Temp', 'Avg Temp','High Temp']);

plt.savefig("Images/daily_normals.png")
plt.show()
```
